```python
import pandas as pd
import statsmodels.api as sm
```

```python
data = pd.read_csv("ApeCoin.csv")
data = data.set_index('Date')
data
```

```python
df = data['Close'].pct_change(periods=2)*100
df
```

```python
df = df.rename("Today")
df
```

```python
df = df.reset_index()
df
```

```python
for i in range (1,6):
    df['lag'+str(i)]=df['Today'].shift(i)
```

```python
df
```

```python
df['Volume'] = data.Volume.shift(1).values/1000_000_000
```

```python
df
```

```python
df = df.dropna()
```

```python
df
```

```python
df['Direction'] = [1 if i>0 else 0 for i in df['Today']]
```

```python
df
```

```python
df = sm.add_constant(df)
```

```python
df
```

```python
x = df[["const","lag1",'lag2','lag3','lag4','lag5','Volume']]
y = df.Direction
```

```python
model = sm.Logit(y,x)
```

```python
result = model.fit()
```

```python
result.summary()
```

```python
prediction = result.predict(x)
```

```python
prediction
```

```python
def c_matrix(act,pred):
    predtrans = ['Up' if i > 0.5 else 'Down' for i in pred]
    actuals = ['Up' if i > 0 else 'Down' for i in act]
    c_matrix = pd.crosstab(pd.Series(actuals),
                                  pd.Series(predtrans),
                                  rownames=['Actual'],
                                  colnames=['Predicted'])
    return c_matrix
```

# Definition
### Bear or Bull only happens in the following situation:
1. The price decreased/increased by 20%
2. This trend continued for 3 months (90 days) --> Meaning no new turns during these sessions

```python
import pandas as pd

import pandas as pd
import numpy as np 
import pickle 
import time 
import random
import sklearn

import matplotlib.pyplot as plt

import seaborn as sns
sns.set(rc={'figure.figsize':(10,6)})
custom_colors = ["#4e89ae", "#c56183","#ed6663","#ffa372"]

import networkx as nx
import plotly.express as px 
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

### ApeCoin

```python
#hasn't fullfilled the condition --> not once has it passed the 90 days condition
#However, since 2022-06-14 there were 4 times where it increased 20%
#2022-06-14 (20.23%) 2022-06-19 (24.59%) 2022-07-17 (25.84%) 2022-07-18 (22.33%)

raw_data=pd.read_csv('ApeCoin.csv')
raw_data.Date = pd.to_datetime(raw_data.Date)
raw_data["prevClose"] = raw_data.Close.shift(1)
raw_data['change'] = raw_data[['Open','prevClose']].pct_change(axis=1)['prevClose']
result = raw_data.loc[abs(raw_data['change']) > 0.20]
result.loc[:,('change')] = result.loc[:,('change')].multiply(100).round(2).astype(str) + "%"

def date_format(df):
    if isinstance(df, pd.DataFrame):
        if 'Date' in df.columns:
            df.Date = pd.to_datetime(df.Date)
            df[['Year','Month','Day']] = df.Date.apply(lambda x: pd.Series(x.strftime("%Y-%m-%d").split("-")))
    else:
        print("Invalid entry")
    return df
data=date_format(raw_data)
data['Close']=data['Close'].astype(float)
data['Open']=data['Open'].astype(float)
data['High']=data['High'].astype(float)
data['Low']=data['Low'].astype(float)
data['Year']=data['Year'].astype(int)
data['Month']=data['Month'].astype(int)
data['Day']=data['Day'].astype(int)

fig = go.Figure(data=go.Ohlc(x=data['Date'],
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close']))

fig.update_layout(template='plotly_dark',
    title='Price over the years',
    xaxis_title="Year",
    yaxis_title='Price',title_x=0.5,
    shapes = [dict(x0='2022-06-14', x1='2022-06-14', y0=0, y1=1, xref='x', yref='paper',line_width=2)],
    annotations=[dict(x='2022-06-14', y=0.05, xref='x', yref='paper',showarrow=False, xanchor='left', text='bull')]
)

fig.update_yaxes( # the y-axis is in dollars
    tickprefix="$", showgrid=False
)

fig.update_xaxes(
     showgrid=True
)

fig.show()
```
